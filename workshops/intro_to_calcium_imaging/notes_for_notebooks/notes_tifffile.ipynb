Reading metadata

```python
with tifffile.TiffFile("../data/data_endoscope.tif") as tif:
    for tag in tif.pages[0].tags.values():
        print(f"  Tag: {tag.name}")
```

```python
import tifffile
import numpy as np

with tifffile.TiffFile("../data/data_endoscope.tif") as tif:
    frames = [page.asarray() for page in tif.pages]

stacked_frames = np.stack(frames, axis=0)
median_image = np.median(stacked_frames, axis=0)
```

```python
plt.imshow(median_image)
```

```python
plt.imshow(np.mean(stacked_frames, axis=0))
```
