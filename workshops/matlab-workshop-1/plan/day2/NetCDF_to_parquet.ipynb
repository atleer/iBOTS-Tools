```python
import xarray as xr
import pandas as pd
dataset = xr.load_dataset('data/steinmetz_2016-12-14_Cori.nc')
dataset = dataset.where(dataset.active_trials==1 , drop=True)
dataset
```

```python
wheel_df = dataset[["wheel","response_type","feedback_type","gocue","response_time","feedback_time"]].to_dataframe().reset_index()
wheel_df_indexed = wheel_df.set_index(['time', 'trial'])
#wheel_df = wheel_df_indexed["wheel"].unstack()
wheel_df.rename(columns={'wheel':'wheel_speed'},inplace=True)
wheel_df.to_parquet('data/steinmetz_2016-12-14_Cori.parquet')
wheel_df
```
